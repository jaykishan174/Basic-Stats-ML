# Statistics In Python
This notebook will present some key statstics concepts using python

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import seaborn as sns
```

## Measure of Central Tendency

* Mean
* Median
* Mode

```
d = {'Name':pd.Series(['Tom','James','Ricky','Vin','Steve','Smith','Jack',
   'Lee','Chanchal','Gasper','Naviya','Andres']),
   'Age':pd.Series([25,26,25,23,30,29,23,34,40,30,51,46]),
   'Rating':pd.Series([4.23,3.24,3.98,2.56,3.20,4.6,3.8,3.78,2.98,4.80,4.10,3.65])}

#Create a DataFrame
df = pd.DataFrame(d)
print ("Mean Values in the Distribution")
print (df.mean())
print ("*******************************")
print ("Median Values in the Distribution")
print (df.median())
print ("Mode Values in the Distribution")
print (df.mode())
```

## Measure of Variance
1. Standard Deviation
2. Skewness
3. Kurtosis

```
print ("Std in the Distribution")
print (df.std())
print ("Skewness in the Distribution")
print (df.skew())
print ("Kurtosis in the Distribution")
print (df.kurtosis())
```

## Normal Distribution
The normal distribution is a form presenting data by arranging the probability distribution of each value in the data.Most values remain around the mean value making the arrangement symmetric.

```
mu, sigma = 0.5, 0.1
s = np.random.normal(mu, sigma, 10000)

# Create the bins and histogram
#count, bins, ignored = plt.hist(s, 20, norm)

# Plot the distribution curve
sns.histplot(s,kde= True)
```

The binomial distribution model deals with finding the probability of success of an event which has only two possible outcomes in a series of experiments.

## Poisson Distribution
A Poisson distribution is a distribution which shows the likely number of times that an event will occur within a pre-determined period of time. It is used for independent events which occur at a constant rate within a given interval of time. The Poisson distribution is a discrete function, meaning that the event can only be measured as occurring or not as occurring, meaning the variable can only be measured in whole numbers.

```
from scipy.stats import poisson
data_binom = poisson.rvs(mu=4, size=10000)
ax = sns.distplot(data_binom,
                  kde=True,
                  color='green',
                  hist_kws={"linewidth": 25,'alpha':1})
ax.set(xlabel='Poisson', ylabel='Frequency')
```

## Central Limit Theorm
The central limit theorem states that if you have a population with mean μ and standard deviation σ and take sufficiently large random samples from the population with replacementtext annotation indicator, then the distribution of the sample means will be approximately normally distributed.

```
population=np.random.randint(0,100,100)
```

```
population
```

```
from random import sample
mean_list=[]
for i in range(0,10000):
    a= np.random.choice(population,10).mean()
    mean_list.append(a)
```

```
mean_list= np.array(mean_list)
```

```
sns.histplot(mean_list, kde=True)
```

```
pop_mean= population.mean()
sample_mean= mean_list.mean()
print("Population Mean is:")
print(pop_mean)
print("Sample Mean is:")
print(sample_mean)
```

```
from math import sqrt
pop_std= population.std()
sample_std= mean_list.std() / (sqrt(10))
print("Population STD is:")
print(pop_std)
print("Sample STD is:")
print(sample_std)
```

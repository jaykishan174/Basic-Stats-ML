```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

### Functional Plot

```python
x= np.linspace(0,10,100)

plt.plot(x, np.sin(x))
plt.show()
```

### Plot Formatting

```python
x= np.linspace(0,10,100)

plt.plot(x, np.sin(x))
plt.title("Title")
plt.xlabel("X Axis Title")
plt.ylabel("Y Axis Title")
plt.show()
```

### Subplot

```python
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))
plt.title("Sin Curve")
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
plt.title("Cos Curve")
plt.show()
```

### Using Figure Objects

```python
fig= plt.figure(figsize=(10,10))
ax = fig.add_axes([0.1,0.1,0.9,0.9])
ax.set_xlabel("Days")
ax.set_ylabel("Sin Days")
ax.set_title("This is Title")
ax.plot(x, np.sin(x))
```

```python
fig,ax= plt.subplots(figsize=(10,10))
ax = fig.add_axes([0.1,0.1,0.9,0.9])
ax.set_xlabel("Days")
ax.set_ylabel("Sin Days")
ax.set_title("This is Title")
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
```

### Subplots

```python
fig,ax= plt.subplots(figsize=(10,10), nrows=2, ncols=1)
#ax = fig.add_axes([0.1,0.1,0.9,0.9])
ax[0].set_xlabel("Days")
ax[0].set_ylabel("Sin Days")
ax[0].set_title("X vs Sin x")
ax[0].plot(x, np.sin(x), color='red', ls='-.', marker='^')
#ax.plot(x, np.cos(x))
ax[1].set_xlabel("Days")
ax[1].set_ylabel("Cos Days")
ax[1].set_title("X vs cos x")
ax[1].plot(x, np.cos(x), color='navy', marker='o', ls='-.')
```

### Inserting text in graphs

```python
fig,ax= plt.subplots(figsize=(10,10))
#ax = fig.add_axes([0.1,0.1,0.9,0.9])
ax.set_xlabel("Days")
ax.set_ylabel("Sin Days")
ax.set_title("X vs Sin x")
ax.text(2,-0.75, "Text is here")
ax.plot(x, np.sin(x), color='red', ls='-.', marker='^')
```

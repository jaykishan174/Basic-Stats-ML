# Seaborn Tutorial

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
crash_df = sns.load_dataset('car_crashes')
```

```python
crash_df.head()
```

### Distribution Plot

#### To find distribution of numerical columns

```python
sns.distplot(crash_df['not_distracted'])    ## Plots histogram and kernel density estimation
```

```python
sns.distplot(crash_df['not_distracted'], kde=False)   ## to turn off kernel desnity curve
```

```python
sns.distplot(crash_df['not_distracted'], bins=25)   ## Number of bins
```

### Joint Plots

#### To find relation between two numerical columns

```python
sns.jointplot(x= 'speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.jointplot(x= 'speeding', y='alcohol', data=crash_df, kind='kde')
```

```python
sns.jointplot(x= 'speeding', y='alcohol', data=crash_df, kind='hex')
```

### KDE Plot

#### To find kernel density distribution of Numerical columns

```python
sns.kdeplot(crash_df['alcohol'])
```

### Pair Plot

```python
sns.pairplot(crash_df)
```

```python
tips_df= sns.load_dataset('tips')
tips_df.head()
```

```python
sns.pairplot(tips_df, hue='sex', palette="Blues")   ## Plots Numerical values, hues are categorical
```

### Rug Plot

```python
sns.rugplot(tips_df['tip'])   ### Where there is more distribution, there is more density
```

### Styling

```python
sns.set_style('dark')
plt.figure(figsize=(10,10))
sns.jointplot(x= 'speeding', y='alcohol', data=crash_df, kind='reg')
```

### Barplot

#### Works with categorical and numerical data

```python
sns.barplot(x='sex', y='total_bill', data=tips_df)   ## by default estimator is mean
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.sum) 
```

### Count data plot

#### Works with categorical data

```python
sns.countplot(x='sex', data=tips_df)
```

### Box Plot

#### Works with categorical and numerical data

```python
sns.boxplot(x='day', y='total_bill', data=tips_df)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')  ## hue of categorical variable
```

### Violin Plot ( Box Plot with KDE curve)

```python
sns.violinplot(x='day', y='total_bill', data=tips_df)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)  ## Combines two hues
```

### Strip Plot

#### Works with categorical and numerical data

```python
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)
```

### Palette

```python
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True, palette='magma')
```

### Heatmap

```python
crash_mx= crash_df.corr()
sns.heatmap(crash_mx, annot=True, cmap='Blues')
```

### Pair Grid ( To customize pair plots)

```python
iris= sns.load_dataset('iris')
```

```python
iris_g= sns.PairGrid(iris, hue= 'species')
iris_g.map(plt.scatter)
```

### Facet Grid

#### To distribute charts among categorical variable

```python
tips_fg= sns.FacetGrid(tips_df, row='smoker', col='time')
tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

```python
tips_fg= sns.FacetGrid(tips_df, hue='smoker', col='time')
tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

### Regression Plot

```python
sns.lmplot(x='total_bill', y='tip', data=tips_df)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips_df, hue='sex', markers=['^','o'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips_df, col='sex', row='time', markers=['^'])
```

```python
sns.scatterplot(x='total_bill', y='tip',data=tips_df, hue='sex')
```

```python
sns.lineplot(x='total_bill', y='tip',data=tips_df, hue='sex')
```
